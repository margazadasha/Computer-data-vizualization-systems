#Завдання 1. Побудувати рекурентну нейронну мережу для прогнозування слів.

```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.layers import Dense, LSTM, SimpleRNN, Dropout, Embedding
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.utils import to_categorical

text = """
Ще не вмерла України, ні слава, ні воля,
Ще нам, браття молодії, усміхнеться доля.
Згинуть наші вороженьки, як роса на сонці,
Запануєм і ми, браття, у своїй сторонці.

Душу й тіло ми положим за нашу свободу,
І покажем, що ми, браття, козацького роду.
"""

# Обробка тексту
text = text.replace('\ufeff', '')

maxWordsCount = 700
tokenizer = Tokenizer(num_words=maxWordsCount, filters='!-"-$#&%;()*+,-./:;<=>?@[\\]^_`{|}~\t\n\r""',
                      lower=True, split=' ', char_level=False)
tokenizer.fit_on_texts([text])

# Перетворення тексту в числову послідовність
data = tokenizer.texts_to_sequences([text])
res = np.array(data[0])

input_words = 4
m = res.shape[0] - input_words

# Формування навчальних вибірок
X = np.array([res[i:i + input_words] for i in range(m)])
Y = to_categorical(res[input_words:], num_classes=maxWordsCount)

# Моделі RNN та LSTM
model = Sequential()
model.add(Embedding(maxWordsCount, 200, input_length=input_words))
model.add(SimpleRNN(128, activation='tanh', return_sequences=True))
model.add(LSTM(128, activation='tanh'))
model.add(Dropout(0.2))
model.add(Dense(128, activation='relu'))
model.add(Dense(maxWordsCount, activation='softmax'))

model.compile(loss='categorical_crossentropy', metrics=['accuracy'], optimizer='adam')

# Навчання моделі
history = model.fit(X, Y, batch_size=32, epochs=100)

# Побудова графіків втрат і точності
plt.figure(figsize=(10, 5))

# Побудова графіку втрат
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Loss', color='lightslategrey' )
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.title('Графік функції втрат')

# Побудова графіку точності
plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label='Accuracy', color='coral')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.title('Графік точності')

plt.show()

model.summary()

# Функція для генерації тексту
def generatePhrase(text, str_length=15):
  res = text
  data = tokenizer.texts_to_sequences([text])[0]
  for i in range(str_length):
    x = data[i: i + input_words]
    input = np.expand_dims(x, axis=0)

    pred = model.predict(input)
    indx = pred.argmax(axis=1)[0]
    data.append(indx)

    # Додавання слова
    res += " " + tokenizer.index_word.get(indx, '')

  return res

# Генерація тексту
res = generatePhrase("Ще не вмерла України")
print(res)
```

```python
import keras_tuner as kt

text = """
Ще не вмерла України, ні слава, ні воля,
Ще нам, браття молодії, усміхнеться доля.
Згинуть наші вороженьки, як роса на сонці,
Запануєм і ми, браття, у своїй сторонці.

Душу й тіло ми положим за нашу свободу,
І покажем, що ми, браття, козацького роду.
"""

# Обробка тексту
text = text.replace('\ufeff', '')

maxWordsCount = 700
tokenizer = Tokenizer(num_words=maxWordsCount, filters='!-"-$#&%;()*+,-./:;<=>?@[\\]^_`{|}~\t\n\r""',
                      lower=True, split=' ', char_level=False)
tokenizer.fit_on_texts([text])

# Перетворення тексту в числову послідовність
data = tokenizer.texts_to_sequences([text])
res = np.array(data[0])

input_words = 4
m = res.shape[0] - input_words

# Формування навчальних вибірок
X = np.array([res[i:i + input_words] for i in range(m)])
Y = to_categorical(res[input_words:], num_classes=maxWordsCount)

# Функція для моделі HyperModel
def generate_model(hp):
  model = Sequential()
  model.add(Embedding(maxWordsCount, hp.Int('embedding_dim', 12, 512, step=32)))
  model.add(SimpleRNN(hp.Int('rnn_units', 64, 256, step=32), activation='tanh', return_sequences=True))
  model.add(LSTM(hp.Int('lstm_units', 64, 256, step=32), activation='tanh'))
  model.add(Dropout(hp.Float('dropout_rate', 0.1, 0.4, step=0.1)))
  model.add(Dense(hp.Int('dense_units', 64, 256, step=32), activation='relu'))
  model.add(Dense(maxWordsCount, activation='softmax'))

  model.compile(
      loss='categorical_crossentropy',
      metrics=['accuracy'],
      optimizer=hp.Choice('optimizer', ['adam', 'rmsprop'])
  )
  return model

# Використання Keras Tuner
tuner = kt.RandomSearch(
    generate_model,
    objective='accuracy',
    max_trials=15,
    executions_per_trial=1,
    directory='my_tuning',
    project_name='text_prediction'
)

# Запуск пошуку гіперпараметрів
tuner.search(X, Y, epochs=10, batch_size=32)

# Підбір параметрів
best_hps = tuner.get_best_hyperparameters(num_trials=1)[0]
print(f"""
Найкращі параметри:
Embedding Dimension: {best_hps.get('embedding_dim')}
RNN Units: {best_hps.get('rnn_units')}
LSTM Units: {best_hps.get('lstm_units')}
Dropout Rate: {best_hps.get('dropout_rate')}
Dense Units: {best_hps.get('dense_units')}
Optimizer: {best_hps.get('optimizer')}
""")

# Побудова моделі з підібраними найкращими параметрами
best_model = tuner.hypermodel.build(best_hps)

# Навчання моделі
history = best_model.fit(X, Y, batch_size=32, epochs=100)

# Побудова графіків втрат і точності
plt.figure(figsize=(10, 5))

# Побудова графіку втрат
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Loss', color='powderblue' )
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.title('Графік функції втрат')

# Побудова графіку точності
plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label='Accuracy', color='plum')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.title('Графік точності')

plt.show()

model.summary()

# Функція для генерації тексту
def generatePhrase(text, str_length=15):
  res = text
  data = tokenizer.texts_to_sequences([text])[0]
  for i in range(str_length):
    x = data[i: i + input_words]
    input = np.expand_dims(x, axis=0)

    pred = best_model.predict(input)
    indx = pred.argmax(axis=1)[0]
    data.append(indx)

    # Додавання слова
    res += " " + tokenizer.index_word.get(indx, '')

  return res

# Генерація тексту
res = generatePhrase("Ще не вмерла України")
print(res)
```

**Висновок:** У даному завданні була побудована рекурентна нейронна мережа для прогнозування слів у тексті. Модель була спроектована з використанням рекурентних шарів, зокрема SimpleRNN та LSTM, що дозволяє ефективно обробляти послідовні дані та захоплювати залежності між словами в тексті.

Підбір архітектури та гіперпараметрів моделі був здійснений за допомогою Keras Tuner, що дозволило оптимізувати параметри, такі як розмір вбудованих шарів (embedding), кількість нейронів у рекурентних шарах, розмір шари LSTM, коефіцієнт відсічення (dropout) та інші. В результаті пошуку гіперпараметрів вдалося отримати модель з оптимальними параметрами, що дозволило досягти дуже гарних результатів у прогнозуванні слів.
